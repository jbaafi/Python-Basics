```python
# missing values in a dataset. How do we handle such
```

```python
missing = np.nan
```

```python
series_obj = Series(["row1", "row2", "row3",missing, "row5", "row6", missing])
series_obj
```

```python
series_obj.isnull()
```

```python
DF_obj.isnull()
```

```python
DF_obj2 = DataFrame(np.random.randn(36).reshape(6, 6))
DF_obj2
```

```python
DF_obj2.ix[3:5, 0] = missing
DF_obj2.ix[1:4, 5] = missing
DF_obj2
```

```python
filled_DF = DF_obj2.fillna(0)
filled_DF
```

```python
DF_obj2 = DataFrame(np.random.randn(36).reshape(6, 6))
DF_obj2.ix[3:5, 0] = missing
DF_obj2.ix[1:4, 5] = missing
DF_obj2
```

```python
#finding which rows have missing values in the dataset
DF_obj2.isnull().sum()
```

```python
#drops all the rows that have missng values and return only rows that have no missing values
DF_obj2.dropna()
```

```python
#drops all the colums that have missng values and return only colums that have no missing values
DF_obj2.dropna(axis=1)
```

```python
#Duplicates
DF_obj3 = DataFrame({"person" : [1, 1, 2, 2, 3, 3, 3], "age": ["a", "a", "b", "b", "c", "c", "c"],
                     "death": ["A", "A", "B", "B", "C", "C", "C"]})
DF_obj3
```

```python
DF_obj3.duplicated()
```

```python
# how to drop duplicates
DF_obj3.drop_duplicates()
```

```python
# Concatenate and transform data ie. joining two or more dataframes together to get a single data frame
DF_obj4 = DataFrame(np.random.randn(36).reshape(6, 6))
DF_obj4
```

```python
DF_obj5 = DataFrame(np.random.randn(15).reshape(5, 3))
DF_obj5
```

```python
pd.concat([DF_obj4, DF_obj5], axis=1)
```

```python
#droping along rows 
DF_obj4.drop([0, 2, 4])
```

```python
#droping along columns
DF_obj4.drop([0, 2, 4], axis = 1)
```
